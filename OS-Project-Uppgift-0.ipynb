```python
import pandas as pd
from pathlib import Path  


os_data = pd.read_csv("../Projekt-OS/Data/athlete_events.csv")
```

### OS Projekt Uppgift O

```python
def data_printer(data):
    for i in range(0, len(data), 5):
        print(f"{", " .join(data[i: i+5])}\n")
```

```python
def data_fetcher(data, typ):
    fetch_list = []
    for i in os_data[typ]:
        if i not in fetch_list:
            fetch_list.append(i)
    return fetch_list
```

```python
# A
noc_list = data_fetcher(os_data, "NOC")


print (f"Det finns: {len(noc_list)} länder i tabellen")
```

```python
# B

print("Länder i OS")
print("-------------------\n")
data_printer(noc_list)
```

```python
# C 

os_data.dtypes
sport_list = data_fetcher(os_data, "Sport")


print("Sporter i OS")
print("-------------------\n")
data_printer(sport_list)
```

```python
# D

medal_list = data_fetcher(os_data, "Medal")
medal_list.pop(0)
print("Medaljer i OS")
print("-------------")
data_printer(medal_list)
```

```python
# g

os_sex_data = os_data["Sex"].value_counts().rename({"M" : "Male", "F": "Female"})   

os_sex_data = pd.DataFrame(os_sex_data)

import seaborn as sns
import matplotlib.pyplot as plt
 

sns.barplot(x = 'Sex',
            y = 'count',
            data = os_sex_data,
            )
 
plt.show()
```

```python
# h
top_10_countries = os_data[os_data['Medal'].notnull()]
top_10_countries = os_data['NOC'].value_counts()
top_10_countries.head(10)
```

```python
# h
import matplotlib.pyplot as plt
import numpy as np
top_10_countries.head(10).plot(kind='bar',
                            color= plt.cm.Paired(np.arange(len(top_10_countries.head(10)))),
                            title=" Top 10 Countries With Most Medals")
plt.xlabel("Country")
plt.ylabel("Total Medals")

plt.show()
```

```python
top_countries_gold = os_data[os_data['Medal'] == 'Gold']['NOC'].value_counts()
top_countries_silver = os_data[os_data['Medal'] == 'Silver']['NOC'].value_counts()
top_countries_bronze = os_data[os_data['Medal'] == 'Bronze']['NOC'].value_counts()

print(f"Top 3 countries in Gold medal: {top_countries_gold.head(3)}")
print(f"Top 3 countries in Silver medal: {top_countries_silver.head(3)}")
print(f"Top 3 countries in Bronze medal: {top_countries_bronze.head(3)}")
```
