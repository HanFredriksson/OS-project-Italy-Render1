```python
def data_printer(data):
    for i in range(0, len(data), 5):
        print(f"{', ' .join(data[i: i+5])}\n")
```

```python
def data_fetcher(data, typ):
    fetch_list = []
    for i in os_data[typ]:
        if i not in fetch_list:
            fetch_list.append(i)
    return fetch_list
```

```python
# h
top_10_countries = os_data[os_data['Medal'].notnull()]
top_10_countries = os_data['NOC'].value_counts()
top_10_countries.head(10)
```

```python
# h
import matplotlib.pyplot as plt
import numpy as np
top_10_countries.head(10).plot(kind='bar',
                            color= plt.cm.Paired(np.arange(len(top_10_countries.head(10)))),
                            title=" Top 10 Countries With Most Medals")
plt.xlabel("Country")
plt.ylabel("Total Medals")

plt.show()
```

```python
# i
def diff_medal():
    top_countries_gold = os_data[os_data['Medal'] == 'Gold']['NOC'].value_counts().head(3)
    top_countries_silver = os_data[os_data['Medal'] == 'Silver']['NOC'].value_counts().head(3)
    top_countries_bronze = os_data[os_data['Medal'] == 'Bronze']['NOC'].value_counts().head(3)
    
    print(f"Top 3 countries in Gold medal: {top_countries_gold.head(3)}")
    print(f"Top 3 countries in Silver medal: {top_countries_silver.head(3)}")
    print(f"Top 3 countries in Bronze medal: {top_countries_bronze.head(3)}")
    
    medals = pd.DataFrame({
        'Gold': top_countries_gold, 
        'Silver': top_countries_silver, 
        'Bronze': top_countries_bronze
    })    
    
    return medals
```

```python
medals = diff_medal()
medals.plot(kind='bar',color= ['Gold', 'silver', 'chocolate'],title= 'Top 3 countries Each Medal Category', fig='grad')
plt.xlabel("Country")
plt.ylabel("Medal Count")
plt.show()
```

#### * In this anlysis we can see that the USA is the top country in all types medals(Gold, Silver, Bronze), and Germany is the second top country in both Gold and Silver medals, while the ussr(URS) holds the second position in Bronze medals.

```python
italy_os = pd.read_csv("../Projekt-OS/Data/italy-OS-data.csv",index_col=0)
```

```python
import hashlib as hl
hashes = italy_os['Name'].apply(lambda x: hl.sha256(x.encode()).hexdigest())
```

```python
italy_os.insert(2,'Name_Hash', hashes)
italy_os.head(3)
```

```python
comparison = (hashes == italy_os['Name_Hash']).all()
comparison
```

```python
# 1
italy_os.fillna(0,inplace=True)
filter_data = italy_os[italy_os['Medal'] !=0]
```

```python
# 1
medal_count = filter_data.groupby('Sport')['Medal'].count().reset_index()
medal_count_sorted = medal_count[medal_count['Medal']> 100]
print(medal_count_sorted.head())

ax = medal_count_sorted.plot(kind='bar', x= 'Sport', y='Medal',
                        title= 'Sports with more then 100 medals',
                        ylabel='Number of medals',
                        color= plt.cm.Paired(np.linspace(0,1,len(medal_count_sorted))))

plt.tight_layout()
plt.show()
```
